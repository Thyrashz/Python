```python
from random import random
```

```python
def collision(persone):
    compleanni = [round(random() * 364) + 1 for _ in persone]
    if len(set(compleanni)) != len(compleanni):
        return True
    else:
        return False
```

```python
collision(classe)
```

```python
def prob(numero_persone, numero_prove = 1000):
    classe = range(numero_persone)
    esperimenti = [collision(classe) for x in range(numero_prove)]
    return sum(esperimenti) / len(esperimenti) * 100
```

```python
prob(23, 20)
```

```python
lista = range(50)
```

```python
lista_probabilità = [prob(x) for x in lista]
```
